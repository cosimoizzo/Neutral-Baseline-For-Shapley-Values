```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

N_sim = 100
```

```python
# Absolute logits
for i in range(N_sim):
    if i == 0:
        df_neutral_abs = pd.read_csv("results/result_abs_neutral_mc_" + str(i) + ".csv")
        df_0_abs = pd.read_csv("results/result_abs_0_mc_" + str(i) + ".csv")
        df_maxdist_abs = pd.read_csv("results/result_abs_maxdist_mc_" + str(i) + ".csv")
        df_pdata_abs = pd.read_csv("results/result_abs_pdata_mc_" + str(i) + ".csv")
    else:
        df_neutral_abs = df_neutral_abs.append(pd.read_csv("results/result_abs_neutral_mc_" + str(i) + ".csv"), ignore_index=True)
        df_0_abs = df_0_abs.append(pd.read_csv("results/result_abs_0_mc_" + str(i) + ".csv"), ignore_index=True)
        df_maxdist_abs = df_maxdist_abs.append(pd.read_csv("results/result_abs_maxdist_mc_" + str(i) + ".csv"), ignore_index=True)
        df_pdata_abs = df_pdata_abs.append(pd.read_csv("results/result_abs_pdata_mc_" + str(i) + ".csv"), ignore_index=True)
    
df_neutral_abs = df_neutral_abs.drop(['Unnamed: 0'], axis=1)
df_0_abs = df_0_abs.drop(['Unnamed: 0'], axis=1)
df_maxdist_abs = df_maxdist_abs.drop(['Unnamed: 0'], axis=1)
df_pdata_abs = df_pdata_abs.drop(['Unnamed: 0'], axis=1)

# Chart
fig = plt.figure(figsize=(35,6))
ax = fig.add_subplot(111) 
gs = fig.add_gridspec(1, 5)
ax11 = fig.add_subplot(gs[0, 0])
ax12 = fig.add_subplot(gs[0, 1])
ax13 = fig.add_subplot(gs[0, 2])
ax14 = fig.add_subplot(gs[0, 3])
#ax15 = fig.add_subplot(gs[0, 4])

x_axis = list(range(1,np.size(df_0_abs,1)+1))

bottom = 0
top = 14
fontsize = 8

fontsize = 24
minus_font = 4

ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)
# Set common labels
ax.set_xlabel("number of features removed", fontsize = fontsize, labelpad=18)

ax11.plot(x_axis, df_0_abs[(df_0_abs!=np.inf)&df_0_abs.notnull()].mean(), 'r', linewidth=1.5)
ax11 = df_0_abs.plot.box(grid='True', title='zero/average', ax=ax11, sym='', fontsize=fontsize-1)
ax11.set_ylabel("expected absolute logits", fontsize=fontsize, labelpad=12)
ax11.title.set_fontsize(fontsize)
ax11.set_ylim((bottom, top)) 
every_nth = 1
for n, label in enumerate(ax11.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)
for n, label in enumerate(ax11.yaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)
        
ax12.plot(x_axis, df_maxdist_abs[(df_maxdist_abs!=np.inf)&df_maxdist_abs.notnull()].mean(), 'r', linewidth=1.5)
ax12 = df_maxdist_abs.plot.box(grid='True', title='mdb', ax=ax12, sym='', fontsize=fontsize-1)
ax12.title.set_fontsize(fontsize)
ax12.set_ylim((bottom, top)) 
ax12.set_yticklabels([])
for n, label in enumerate(ax12.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)

ax13.plot(x_axis, df_pdata_abs[(df_pdata_abs!=np.inf)&df_pdata_abs.notnull()].mean(), 'r', linewidth=1.5)
ax13 = df_pdata_abs.plot.box(grid='True', title='p$_{X}$', ax=ax13, sym='', fontsize=fontsize-1)
ax13.title.set_fontsize(fontsize)
ax13.set_yticklabels([])
ax13.set_ylim((bottom, top)) 
for n, label in enumerate(ax13.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)

ax14.plot(x_axis, df_neutral_abs[(df_neutral_abs!=np.inf)&df_neutral_abs.notnull()].mean(), 'r', linewidth=1.5)
ax14 = df_neutral_abs.plot.box(grid='True', title='neutral$_{0.5}$', ax=ax14, sym='', fontsize=fontsize-1)
ax14.title.set_fontsize(fontsize)
ax14.set_yticklabels([])
ax14.set_ylim((bottom, top)) 
for n, label in enumerate(ax14.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)
        
fig.savefig('plots/logits_synth.pdf', bbox_inches = "tight")
```

```python
from sklearn import metrics
def assign_score_to_method(method_perf):
    x = np.linspace(0,len(method_perf)-1,len(method_perf))[::-1]
    _out = metrics.auc(x, method_perf)
    return _out
```

```python
folder_results = "results/"
csv_name = "results_ROAR_mc_"
names = list()
performances_all = list()
keep_count = 0
for n in range(N_sim):
    thiscsv = pd.read_csv(folder_results+csv_name+str(n)+'.csv')
    performances = list()
    if n == 0:
        mean_thiscsv = thiscsv.copy()
        for thiskey in thiscsv.keys():
            if thiskey[-4:] == "perf":
                names.append(thiskey)
                performances.append(assign_score_to_method(thiscsv[thiskey]))
    else:
        mean_thiscsv = mean_thiscsv+thiscsv
        for thiskey in names:
            performances.append(assign_score_to_method(thiscsv[thiskey]))
    performances_all.append(performances)
    keep_count += 1

mean_thiscsv = mean_thiscsv/keep_count

df = pd.DataFrame(performances_all[0], index = names, columns = [0])
for n in range(1,len(performances_all)):
    df = df.join(pd.DataFrame(performances_all[n], index = names, columns = [n]))
df = df.T

df = df.rename(columns={"zero _perf": "zero/average", "neutral05 _perf": "neutral$_{0.5}$", 
                        "max dist _perf": "mdb", "pdata _perf": "p$_{X}$", 
                        "random _perf": "random"})

cols = ['random', 'zero/average', 'mdb', 'p$_{X}$', 'neutral$_{0.5}$']
df = df[cols]
```

```python
fig = plt.figure(figsize=(14,6))
gs = fig.add_gridspec(1, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
x_axis = range(len(mean_thiscsv))

fontsize = 18
ax1.plot(x_axis, mean_thiscsv['random _perf'], label='random', linestyle ='dashed')
ax1.plot(x_axis, mean_thiscsv['zero _perf'], label='zero/average')
ax1.plot(x_axis, mean_thiscsv['max dist _perf'], label='mdb')
ax1.plot(x_axis, mean_thiscsv['pdata _perf'], label='p$_{X}$')
ax1.plot(x_axis, mean_thiscsv['neutral05 _perf'], label='neutral$_{0.5}$')
ax1.set_xlabel("number of features removed", fontsize = fontsize)
ax1.set_ylabel("model performance", fontsize = fontsize)
ax1.legend(loc='upper right',prop={'size': fontsize})
ax1.tick_params(axis="x", labelsize=fontsize)
ax1.tick_params(axis="y", labelsize=fontsize)

ax2 = df.plot.box(grid='True', ax=ax2, sym='', fontsize = fontsize)
plt.xticks(rotation=20)
plt.tight_layout()

fig.savefig('plots/ROAR_synth.pdf')
```
